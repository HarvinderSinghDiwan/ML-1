```python
import pandas as pd
# reading an excel file
output=pd.read_excel('cert.xlsx')
# print all columns name
# output.columns
#   print one columns 
output['STUDENT NA']
```

```python
# Reading a particular sheet from xlsx file
output1=pd.read_excel('cert.xlsx',sheetname='Sheet2')
#printing back bottom 5 lines
output1.tail()
```

```python
#  printing  xlsx file but not using first row as Heading or Column names
output2=pd.read_excel('cert.xlsx',header=None)
#   header as starting from 0 to n === 0 is by default
output2.head()
```

```python
#  making  column as row index number 
output3=pd.read_excel('cert.xlsx',index_col=None)

output3.head()
```

```python
#  reading the file and skip first two rows  
# note it will automatically put the 2nd row as heading columns
output4=pd.read_excel('cert.xlsx',skiprows=2)
output4.head()
```
